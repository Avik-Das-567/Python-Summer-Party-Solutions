## Question 2

Convert the 'order_date' column to a datetime format and create a MultiIndex with 'customer_id' and 'order_date'. Then, calculate the total number of returns per customer for each month. This will provide insights into monthly return patterns for each customer.

```python
import pandas as pd

# Convert order_date to datetime (robust parsing)
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce', infer_datetime_format=True)

# Filter only returns with valid dates
returns_df = df[(df['return_flag'] == True) & (df['order_date'].notna())]

# Set MultiIndex: customer_id + order_date
returns_df = returns_df.set_index(['customer_id', 'order_date']).sort_index()

# Group by customer_id + year-month, and count returns
monthly_returns = (
    returns_df
    .groupby([returns_df.index.get_level_values(0), returns_df.index.get_level_values(1).to_period('M')])
    .size()
    .unstack(fill_value=0)
)

print(monthly_returns)
```
