## Question 1

How many transactions in October 2024 have a customer email ending with a domain other than 'gmail.com', 'yahoo.com', or 'hotmail.com'? This metric will help us identify transactions associated with less common email providers that may indicate emerging risk patterns.

```python
import pandas as pd

# Create fct_transactions dataframe
fct_transactions = pd.DataFrame([
    ["alice@gmail.com", 1, "2024-10-05", 120, 10],
    ["bob@customdomain.com", 2, "2024-10-15", 250.5, 20],
    ["charlie@yahoo.com", 3, "2024-10-20", 75.25, 15],
    ["dana@hotmail.com", 4, "2024-10-25", 100, 30],
    ["eve@biz.org", 5, "2024-10-30", 300, 40],
    ["frank@gmail.com", 6, "2024-11-03", 150.75, 25],
    ["grace@outlook.com", 7, "2024-11-10", None, 50],
    ["ivan@yahoo.com", 8, "2024-11-15", 200, 35],
    ["judy@hotmail.com", 9, "2024-11-21", 250, 45],
    ["ken@domain.net", 10, "2024-11-29", 300, 55],
    ["laura@riskmail.com", 11, "2024-12-02", 100, 80],
    ["mike@securepay.com", 12, "2024-12-03", 180, 85],
    ["nina@trusthub.com", 13, "2024-12-09", 220, 90],
    ["oscar@fintech.com", 14, "2024-12-16", 140, 70],
    ["paula@alertsys.com", 15, "2024-12-23", 260, 95]
], columns=["customer_email", "transaction_id", "transaction_date", "transaction_amount", "fraud_detection_score"])

fct_transactions["transaction_date"] = pd.to_datetime(fct_transactions["transaction_date"])

# Filter for October 2024
october_data = fct_transactions[
    (fct_transactions["transaction_date"].dt.month == 10) &
    (fct_transactions["transaction_date"].dt.year == 2024)
]

# Exclude common domains
exclude_domains = ["gmail.com", "yahoo.com", "hotmail.com"]
october_data = october_data[~october_data["customer_email"].str.endswith(tuple(exclude_domains))]

# Count
count_result = len(october_data)
count_result
```

## Question 2

For transactions occurring in November 2024, what is the average transaction amount, using 0 as a default for any missing values? This calculation will help us detect abnormal transaction amounts that could be related to fraudulent activity.

```python
# Filter for November 2024
nov_data = fct_transactions[
    (fct_transactions["transaction_date"].dt.month == 11) &
    (fct_transactions["transaction_date"].dt.year == 2024)
]

# Replace NaN with 0
nov_data["transaction_amount"] = nov_data["transaction_amount"].fillna(0)

# Calculate average
avg_amount = nov_data["transaction_amount"].mean()
avg_amount
```

## Question 3

Among transactions flagged as 'High' risk in December 2024, which day of the week recorded the highest number of such transactions? This analysis is intended to pinpoint specific days with concentrated high-risk activity and support the development of our preliminary fraud detection score.

```python
# Create dim_risk_flags DataFrame
dim_risk_flags = pd.DataFrame([
    ["Low", 1, 2],
    ["Medium", 2, 7],
    ["High", 3, 11],
    ["High", 4, 12],
    ["High", 5, 13],
    ["Medium", 6, 14],
    ["High", 7, 15],
    ["Low", 8, 1],
    ["Medium", 9, 6],
    ["Low", 10, 3]
], columns=["risk_level", "risk_flag_id", "transaction_id"])

# Join tables
merged = pd.merge(fct_transactions, dim_risk_flags, on="transaction_id")

# Filter for High risk in December 2024
high_dec = merged[
    (merged["risk_level"] == "High") &
    (merged["transaction_date"].dt.month == 12) &
    (merged["transaction_date"].dt.year == 2024)
]

# Count per weekday
weekday_counts = high_dec["transaction_date"].dt.day_name().value_counts()

# Get the weekday with max count
top_day = weekday_counts.idxmax()
top_day
```
