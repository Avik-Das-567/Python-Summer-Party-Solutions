## Question 1

What is the average spending per guest per visit for each park experience type during July 2024? Ensure that park experience types with no recorded transactions are shown with an average spending of 0.0. This analysis helps establish baseline spending differences essential for later segmentation.

```python
import pandas as pd
from io import StringIO

# Sample data
data = """guest_id,visit_date,amount_spent,park_experience_type
1,2024-07-05,50,Attraction
2,2024-07-06,30,Dining
3,2024-07-10,20.5,Retail
4,2024-07-12,40,Entertainment
1,2024-07-15,35,Dining
5,2024-07-20,60,Attraction
6,2024-07-25,25,Retail
1,2024-08-03,55,Attraction
1,2024-08-15,45,Dining
2,2024-08-05,22,Retail
2,2024-08-20,38,Entertainment
7,2024-08-10,15,Character Meet
3,2024-08-25,28,Retail
3,2024-08-27,32,Dining
1,2024-09-02,65,Attraction
8,2024-09-05,50,Retail
9,2024-09-15,40,Dining
10,2024-09-20,70,Entertainment
1,2024-09-25,35,Dining
8,2024-09-28,10,Character Meet
"""

# Load data
df = pd.read_csv(StringIO(data), parse_dates=['visit_date'])

# Extract July 2024 data
july_df = df[
    (df['visit_date'].dt.year == 2024) &
    (df['visit_date'].dt.month == 7)
]

# Get all unique experience types from the entire dataset
all_experiences = df['park_experience_type'].unique()

# Group by experience type and calculate average
july_avg = july_df.groupby('park_experience_type')['amount_spent'].mean().reset_index()

# Reindex with all experience types
july_avg_full = pd.DataFrame({'park_experience_type': all_experiences}).merge(
    july_avg, on='park_experience_type', how='left'
)

# Fill NaN with 0.0
july_avg_full['amount_spent'] = july_avg_full['amount_spent'].fillna(0.0)

# Round for cleaner display
july_avg_full['amount_spent'] = july_avg_full['amount_spent'].round(2)

# Final Output
print("ðŸ“Š Average Spending per Guest per Visit (July 2024):\n")
print(july_avg_full.sort_values(by='amount_spent', ascending=False).to_string(index=False))
```

## Question 2

For guests who visited our parks more than once in August 2024, what is the difference in spending between their first and their last visit? This investigation, using sequential analysis, will reveal any shifts in guest spending behavior over multiple visits.

```python
import pandas as pd
from io import StringIO

# Sample data
data = """guest_id,visit_date,amount_spent,park_experience_type
1,2024-07-05,50,Attraction
2,2024-07-06,30,Dining
3,2024-07-10,20.5,Retail
4,2024-07-12,40,Entertainment
1,2024-07-15,35,Dining
5,2024-07-20,60,Attraction
6,2024-07-25,25,Retail
1,2024-08-03,55,Attraction
1,2024-08-15,45,Dining
2,2024-08-05,22,Retail
2,2024-08-20,38,Entertainment
7,2024-08-10,15,Character Meet
3,2024-08-25,28,Retail
3,2024-08-27,32,Dining
1,2024-09-02,65,Attraction
8,2024-09-05,50,Retail
9,2024-09-15,40,Dining
10,2024-09-20,70,Entertainment
1,2024-09-25,35,Dining
8,2024-09-28,10,Character Meet
"""

# Load data
df = pd.read_csv(StringIO(data), parse_dates=['visit_date'])

# Step 1: Filter for August 2024
august_df = df[
    (df['visit_date'].dt.year == 2024) &
    (df['visit_date'].dt.month == 8)
]

# Step 2: Count visits per guest
visit_counts = august_df['guest_id'].value_counts()
repeat_guests = visit_counts[visit_counts > 1].index

# Step 3: Filter only repeat guests
repeat_df = august_df[august_df['guest_id'].isin(repeat_guests)]

# Step 4: Calculate first and last spending for each guest
def calc_spending_diff(group):
    sorted_group = group.sort_values('visit_date')
    first = sorted_group.iloc[0]['amount_spent']
    last = sorted_group.iloc[-1]['amount_spent']
    return pd.Series({
        'first_visit_spending': first,
        'last_visit_spending': last,
        'spending_difference': round(last - first, 2)
    })

result = repeat_df.groupby('guest_id').apply(calc_spending_diff).reset_index()

# Output
print("ðŸ§¾ Spending Difference for Repeat Guests (August 2024):\n")
print(result.to_string(index=False))
```

## Question 3

In September 2024, how can guests be categorized into distinct spending segments such as Low, Medium, and High based on their total spending? Use the following thresholds for categorization: 
-Low: Includes values from $0 up to, but not including, $50.
-Medium: Includes values from $50 up to, but not including, $100.
-High: Includes values from $100 and above. 
Exclude guests who did not make any purchases in the period.

```python
import pandas as pd
from io import StringIO

# Sample data
data = """guest_id,visit_date,amount_spent,park_experience_type
1,2024-07-05,50,Attraction
2,2024-07-06,30,Dining
3,2024-07-10,20.5,Retail
4,2024-07-12,40,Entertainment
1,2024-07-15,35,Dining
5,2024-07-20,60,Attraction
6,2024-07-25,25,Retail
1,2024-08-03,55,Attraction
1,2024-08-15,45,Dining
2,2024-08-05,22,Retail
2,2024-08-20,38,Entertainment
7,2024-08-10,15,Character Meet
3,2024-08-25,28,Retail
3,2024-08-27,32,Dining
1,2024-09-02,65,Attraction
8,2024-09-05,50,Retail
9,2024-09-15,40,Dining
10,2024-09-20,70,Entertainment
1,2024-09-25,35,Dining
8,2024-09-28,10,Character Meet
"""

# Load data
df = pd.read_csv(StringIO(data), parse_dates=['visit_date'])

# Step 1: Filter for September 2024
sept_df = df[
    (df['visit_date'].dt.year == 2024) &
    (df['visit_date'].dt.month == 9)
]

# Step 2: Calculate total spending per guest
guest_totals = sept_df.groupby('guest_id')['amount_spent'].sum().reset_index()

# Step 3: Exclude guests with zero spending
guest_totals = guest_totals[guest_totals['amount_spent'] > 0]

# Step 4: Categorize spending using precise thresholds
def categorize(amount):
    if 0 <= amount < 50:
        return "Low"
    elif 50 <= amount < 100:
        return "Medium"
    else:
        return "High"

guest_totals['spending_segment'] = guest_totals['amount_spent'].apply(categorize)

# Step 5: Round for display
guest_totals['amount_spent'] = guest_totals['amount_spent'].round(2)

# Step 6: Output
print("ðŸ§¾ Guest Spending Segments (September 2024):\n")
print(guest_totals.sort_values(by='amount_spent', ascending=False).to_string(index=False))
```
