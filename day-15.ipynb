## Question 1

What is the average driver earnings per completed UberPool ride with more than two riders between July 1st and September 30th, 2024? This analysis will help isolate trips that meet specific rider thresholds to understand their impact on driver earnings.

```python
import pandas as pd
from datetime import datetime

# Create the DataFrame
data = [
    [101, 1, 'UberPool', '2024-07-05', 3, 10.5, 22.5],
    [102, 1, 'UberPool', '2024-07-15', 2, 8, 18],
    [103, 2, 'UberPool', '2024-08-10', 4, 15, 35],
    [104, 3, 'UberX', '2024-07-20', 1, 5, 12],
    [105, 2, 'UberPool', '2024-09-01', 3, 12, 30],
    [106, 4, 'UberPool', '2024-09-15', 5, 20, 50],
    [107, 4, 'UberPool', '2024-10-01', 3, 9, 25],
    [108, 5, 'UberPool', '2024-08-25', 4, 11, 28],
    [109, 1, 'UberPool', '2024-09-30', 3, 6, 16],
    [110, 2, 'UberPool', '2024-07-07', 2, 7, 15],
    [111, 3, 'UberPool', '2024-08-05', 4, 13, 32],
    [112, 5, 'UberX', '2024-09-10', 1, 4, 10],
    [113, 6, 'UberPool', '2024-07-30', 3, 22, 45],
    [114, 6, 'UberPool', '2024-08-22', 4, 18, 42],
    [115, 7, 'UberPool', '2024-09-21', 5, 25, 60],
]

columns = ["trip_id", "driver_id", "ride_type", "trip_date", "rider_count", "total_distance", "total_earnings"]
df = pd.DataFrame(data, columns=columns)

# Convert trip_date to datetime
df['trip_date'] = pd.to_datetime(df['trip_date'])

# Filter conditions
start_date = datetime(2024, 7, 1)
end_date = datetime(2024, 9, 30)

filtered_df = df[
    (df['ride_type'] == 'UberPool') &
    (df['rider_count'] > 2) &
    (df['trip_date'] >= start_date) &
    (df['trip_date'] <= end_date)
]

# Calculate average driver earnings
average_earnings = filtered_df['total_earnings'].mean()
average_earnings
```

## Question 2

For completed UberPool rides between July 1st and September 30th, 2024, derive a new column calculating earnings per mile (total_earnings divided by total_distance) and then compute the average earnings per mile for rides with more than two riders. This calculation will reveal efficiency metrics for driver compensation.

```python
# Filter for UberPool rides between the given dates
filtered_df2 = df[
    (df['ride_type'] == 'UberPool') &
    (df['trip_date'] >= start_date) &
    (df['trip_date'] <= end_date)
].copy()

# Calculate earnings per mile
filtered_df2['earnings_per_mile'] = filtered_df2['total_earnings'] / filtered_df2['total_distance']

# Filter for rides with more than two riders
filtered_gt2 = filtered_df2[filtered_df2['rider_count'] > 2]

# Calculate average earnings per mile
avg_earnings_per_mile = filtered_gt2['earnings_per_mile'].mean()
avg_earnings_per_mile
```

## Question 3

Identify the combination of rider count and total distance that results in the highest average driver earnings per UberPool ride between July 1st and September 30th, 2024. This analysis directly recommends optimal trip combination strategies to maximize driver earnings.

```python
# Filter for UberPool rides in the given date range
filtered_df3 = df[
    (df['ride_type'] == 'UberPool') &
    (df['trip_date'] >= start_date) &
    (df['trip_date'] <= end_date)
]

# Group by rider_count and total_distance, compute average earnings
grouped = filtered_df3.groupby(['rider_count', 'total_distance'])['total_earnings'].mean().reset_index()

# Identify the combination with the highest average earnings
max_combination = grouped.loc[grouped['total_earnings'].idxmax()]
max_combination
```
